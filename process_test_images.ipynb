```python
%matplotlib inline
import numpy as np
import cv2
import os
import matplotlib.pylab as plt
import matplotlib.image as mpimg
from six.moves import cPickle as pickle
```

```python
path = 'test_images/'
images = ['{}'.format(path)+image for image in os.listdir(path) if image.endswith('jpg') or image.endswith('png')]
images.sort()
processed = []
```

```python
print(images)
```

```python
image_1 = mpimg.imread(images[0])
image_1 = cv2.resize(image_1, (32,32))
image_1 = cv2.cvtColor(image_1, cv2.COLOR_RGBA2RGB)
image_1 = (image_1*255).astype(np.uint8)
processed.append(image_1)
plt.imshow(image_1);
```

```python
image_2 = mpimg.imread(images[1])
image_2 = image_2[:125,25:150,:]
image_2 = cv2.resize(image_2,(32,32))
processed.append(image_2)
plt.imshow(image_2);
```

```python
image_3 = mpimg.imread(images[2])
image_3 = image_3[:175,75:250,:]
image_3 = cv2.resize(image_3,(32,32))
processed.append(image_3)
plt.imshow(image_3)
```

```python
image_4 = mpimg.imread(images[3])
image_4 = cv2.resize(image_4,(32,32))
image_4 = cv2.cvtColor(image_4, cv2.COLOR_RGBA2RGB)
image_4 = (image_4*255).astype(np.uint8)
processed.append(image_4)
plt.imshow(image_4)
```

```python
image_5 = mpimg.imread(images[4])
image_5 = image_5[:210, 15:265,:]
image_5 = cv2.resize(image_5, (32,32))
processed.append(image_5)
plt.imshow(image_5)
```

```python
image_6 = mpimg.imread(images[5])
image_6 = cv2.resize(image_6, (32,32))
processed.append(image_6)
plt.imshow(image_6)
```

```python
test_data = np.array(processed, dtype=np.uint8)
```

```python
with open('new_test.p', 'wb') as test_file:
    pickle.dump(test_data, test_file)
```
